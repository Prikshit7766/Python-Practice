**Expressions**

2 + 2 is called an expression, which is the most basic kind of 
programming instruction in the language. Expressions consist of values
(such as 2) and operators (such as +), and they can always evaluate (that is, 
reduce) down to a single value.
    


. The ** operator is evaluated first; 
the *, /, //, and % operators are evaluated next, from left to right; and the + and - operators are evaluated last
