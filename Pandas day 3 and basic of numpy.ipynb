# merge 

```python
d3={
    "emp_id":[101,102,103,104],
    "salary":[234,234,434,400],
    "pf":[23,45,67,89],
}
```

```python
d4={
    "emp_id":[101,102,104,105],
    "mobile_no":[2312,123323,12323,123232434],
    "house_nu":[345,354,444,67],
    
}
```

```python
df3=pd.DataFrame(d3)
```

```python
df4=pd.DataFrame(d4)
```

```python
df3
```

```python
df4
```

```python
# some how these two data frame are related to each other
```

```python
# merge and join opeation in pandas
```

```python
# merge
pd.merge(df3,df4) # by default it is performing inner merge operation
```

```python
# it is taking the commen one ---> in emp_id
```

```python
pd.merge(df3,df4 ,how ="left")  #left merge every thing from left
                                 # here left data is df3
    # here it is taking everything from df3 and then performing merge operation
```

```python
pd.merge(df3,df4 ,how ="right") #right  merge every thing from left
                                 # here right data is df4
    # here it is taking everything from df4 and then performing merge operation
```

```python
# specify on waht you are going to do join opeartion ---> that is providing the column name
```

```python
pd.merge(df3,df4 ,how ="right",on="emp_id")
```

```python
# create new column  -- profit flag to store pos or neg based on the profit column
```

```python
def pos_flag(a):
    if a>0:
        return "positive"
    else:
        return "negative"
```

```python
pos_flag(45)
```

```python
# so we can use "apply function"

```

```python
data["profit_fag"]=data["profit"].apply(pos_flag)
```

```python
#or
data["profit_fag_"]=data["profit"]>0
```

```python
data.head()
```

```python
#create one more column where you can store length of customer_name
```

```python
data["len_customer_name"]=data["customer_name"].apply(len)
```

```python
# lets filter out some value 

#categorise the qunatity in 3 segments
'''1 high 
2 medium
3 low'''
```

```python
def quant(a):
    if a<10:
        return "low"
    elif a>10 and a<20 :
        return "medium"
    elif a>20:
        return "high"
```

```python
data["flag_quantit"]=data["quantity"].apply(quant)
```
